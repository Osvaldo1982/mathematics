```python
import sympy as sym
import numpy as np
import pandas as pd
```

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
```

```python
p1 = sym.plot(x*x)
p2 = sym.plot(x)
p1.append(p2[0])
p1
```

```python
sym.plot(x, x**2, x**3, (x, -5, 5))
```

```python
from sympy.plotting import plot3d

plot3d(x*y, (x, -5, 5), (y, -5, 5))
```

```python
plot3d(x*y, -x*y, (x, -5, 5), (y, -5, 5))
```
