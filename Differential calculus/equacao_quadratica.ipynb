## Resumo teórico

Uma equação do segundo grau $ax^2 + bx + c = 0$, para $a \neq 0$, tem como soluçõe reais $x_+$ e $x_-$ dadas pelas seguintes expressões

\begin{equation}
\nonumber x_{+} = \frac{-b + \sqrt{b^2 - 4ac}}{2a} \ \ , \ \ x_{-} = \frac{-b - \sqrt{b^2 - 4ac}}{2a}.
\end{equation}

Este método de encontrar as raízes é válido somente quando o termo dentro da raíz quadrada, comumente chamado de descriminante, for maior que zero. Em geral se representa o descrininante pela letra grega $\Delta$.

\begin{equation}
\nonumber \Delta = b^2 - 4ac.
\end{equation}

* $\Delta >0:$ A equação possui duas raízes reais.
* $\Delta =0:$ A equação possui duas raízes idênticas.
* $\Delta < 0:$ A equação possui raízes complexas.

## O gráfico da parábola

Os principais parâmetros do gráfico de uma parábola são seu vértice e consequentemente suas coordenadas $(x_v,y_v)$, suas raízes $x_1, x_2$ e se a parábola cruza ou toca o eixo x das abscissas. O $x_v$ é a coordenada do eixo das abscissas no qual pertencem ao vértice da parábola e o $y_v$ o valor da ordenada, sendo calculados pelas fórmulas

\begin{equation}
\nonumber x_v = - \frac{b}{2a} \ \ , \ \ y_v = - \frac{\Delta}{4a}.
\end{equation}

O descriminante $(\Delta)$ da equação do segundo grau pode nos informar um pouco sobre a geometria da parábola, abaixo um resumo de como o gráfico se comporta de acordo com valores de $\Delta$:

* $\Delta >0:$ O gráfico corta o eixo x em dois pontos.
* $\Delta =0:$ O gráfico apenas toca o eixo x em um ponto.
* $\Delta < 0:$ O gráfico não corta o eixo x em nenhum ponto.

```python
import math

a = int(input("Insira valores para a: "))
b = int(input("Insira valores para b: "))
c = int(input("Insira valores para c: "))

d = b**2-4*a*c 

if d < 0:
    print ("Esta equação não possui raízes reais")
elif d == 0:
    x = ( -b + math.sqrt(d) )/2*a
    print ("Esta equação possui duas soluções iguais: ", x)
else:
    x1 = (-b+math.sqrt((b**2)-(4*(a*c))))/(2*a)
    x2 = (-b-math.sqrt((b**2)-(4*(a*c))))/(2*a)
    print ("Soluções para esta equação: ", x1, " ou", x2)
```

## Usando numpy

Nesta seção vamos utilizar a biblioteca sympy para encontra raízes de equações do segundo grau. Primeiramente vamos conhecer a função .roots() do numpy. Primeiramente precisamos definir um polinômio $p(x)$ de grau $n$, como mostra a fórmula abaixo

\begin{equation}
\nonumber p(x) a_0 + a_1 x + a_2 x^2 + \ldots + a_{n-1}x^{n-1} + a_n x^n,
\end{equation}

sendo $a_0, a_1, \ldots, a_n$ os coeficientes deste polinômio que também podem ser complexos, mas para os exemplos seguintes iremos utilizar apenas coeficientes reais em polinômios de segunda ordem e encontrar as raízes deste polinômio com a função .roots(), que é o equivalente a solucionar uma equação de segundo grau.

```python
import numpy as np

coeff = [3.2, 2, 1]
np.roots(coeff)
```

## Usando Simpy

Nesta subseção vamos aprender a manipular símbolos em problemas algébricos utilizando a biblioteca sympy. Abaixo um exemplo de como definimos os símbolos que gostaríamos de utilizar em expressões algébricas e funções e a partir daí podemos realizar manipulações simbólicas como derivação e integração de funções, simplificação de expressões, fatoração de binômios e álgebra básica. Vamos aprender sobre três funções .symbols(()), .Eq() e .solve() que requer dois argumentos, o primeiro sendo a equação a ser resolvida e no segundo argumento para qual variável deve ser resolvida a equação.

```python
import sympy as sp
sp.init_printing()

a, b, c, x = sp.symbols(('a', 'b', 'c', 'x'))

# Define uma equação
eq = sp.Eq(a*x**2+b*x+c)

# Resolve uma equação, requer pelo menos dois argumentos.
sp.solve(eq,x)
```

## Exemplo avançado 1

```python
#Iterate these 5 coeficients and plot each line
coefs = [-2, -1, 0, 1, 2]

#set up the plot and 3 subplots (to show the effect of varying each coefficient)
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(18, 6))

#some x values to plot
x = np.linspace(-2, 2, 30)

for idx, val in enumerate([ax1, ax2, ax3]):

    for v in coefs:        
        a, b, c = 1, 0, 0

        if idx == 0:
            a = v
        elif idx == 1:
            b = v
        else:
            c = v

        y = a * (x**2) + (b * x) + c

        val.plot(x, y, label="Coeficient is " + str(coefs[idx]))

        val.axhline(y=0, color='k')
        val.axvline(x=0, color='k')    
        val.grid()
        val.legend(loc='lower center')

plt.show()
```

## Exemplo avançado 2

```python
coefs = np.array([-2, -1, 0, 1, 2])

#set up the plot and 3 subplots (to show the effect of varying each coefficient)
f, axes = plt.subplots(1, 3, sharey=True, figsize=(18, 6))

#some x values to plot
x = np.linspace(-2, 2, 30)

# calculate the y-values, varying each coefficient separately
y_values = [coefs[:, None] * x[None, :] ** 2,                 # vary a
            x[None, :] ** 2 + coefs[:, None] * x[None, :],    # vary b
            x[None, :] ** 2 + coefs[:, None]]                 # vary c

for ax, ys in zip(axes, y_values):
    for c, y in zip(coefs, ys):
        ax.plot(x, y, label="Coefficient is {}".format(c))
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.grid()
    ax.legend()
plt.show()
```
